# Data Exploration and Visualization

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
unknown = pd.read_csv('data/preprocessing_II/unknown.csv')
white = pd.read_csv('data/preprocessing_II/white.csv')
other = pd.read_csv('data/preprocessing_II/other.csv')
asian = pd.read_csv('data/preprocessing_II/asian.csv')
hispanic_latino = pd.read_csv('data/preprocessing_II/hispanic_latino.csv')
black_african_american = pd.read_csv('data/preprocessing_II/black_african_american.csv')
unable_to_obtain = pd.read_csv('data/preprocessing_II/unable_to_obtain.csv')
american_indian_alaska_native = pd.read_csv('data/preprocessing_II/american_indian_alaska_native.csv')
```

## Ethnicity Distribution

```python
ethnicities = np.array(['unknown', 'white', 'other', 'asian', 'hispanic_latino', 'black_african_american', 
                            'unable_to_obtain', 'american_indian_alaska_native'])
ethnicity_distribution = np.array([len(unknown), len(white), len(other), len(asian), 
                            len(hispanic_latino), len(black_african_american), len(unable_to_obtain), 
                            len(american_indian_alaska_native)])

plt.figure(figsize=(10,10))
plt.title('Ethnicity Distribution', fontsize=40)
plt.pie(ethnicity_distribution, labels=ethnicities)
plt.show()
```
